```python
import pandas as pd
import numpy as np
```

```python
nomads_all_merged=pd.read_csv('../data/clean/nomads_all_merged.csv', sep=";")
```

```python
nomads_all_merged
```

```python
# Create the new column using equal-width binning
nomads_all_merged['internet_speed_category'] = pd.cut(
    nomads_all_merged['internet_speed'],
    bins=5,  # Divide the range into 5 equal intervals
    labels=["Very Slow", "Slow", "Moderate", "Fast", "Very Fast"]
)
```

```python
# Min-max normalization (inverted) to scale between 0 and 5
min_val = nomads_all_merged['cost_of_living_value'].min()
max_val = nomads_all_merged['cost_of_living_value'].max()

nomads_all_merged['cost_of_living_score'] = (
    (max_val - nomads_all_merged['cost_of_living_value']) / (max_val - min_val) * 5
).round(1)
```

```python
trip_df_cleaned=pd.read_csv('../data/clean/trip_df_cleaned.csv', sep=";")
```

```python
trip_df_cleaned
```

```python
users_wework=pd.read_csv('../data/clean/users_wework.csv', sep=";")
```

```python
users_wework['coworking_potential_category'] = pd.qcut(
    users_wework['coworking_potential'],
    q=5,
    labels=["Very Low", "Low", "Moderate", "High", "Very High"]
)
```

```python
# Min-max normalization to scale between 0 and 5
min_val = users_wework['coworking_potential'].min()
max_val = users_wework['coworking_potential'].max()

users_wework['coworking_potential_score'] = (
    (users_wework['coworking_potential'] - min_val) / (max_val - min_val) * 5
).round(1)
```

```python
users_wework
```

```python
nomads_all_merged = nomads_all_merged.merge(
    users_wework[['country_code', 'coworking_potential_score']],
    on='country_code',
    how='left'
)

# Fill missing values with 0
nomads_all_merged['coworking_potential_score'] = nomads_all_merged['coworking_potential_score'].fillna(0)
```

```python
nomads_all_merged
```

```python
# Step 1: Create a mask for top 100 ranks
top_100_mask = nomads_all_merged['rank'] <= 100

# Step 2: Get min and max from the filtered top 100
min_rank = nomads_all_merged.loc[top_100_mask, 'rank'].min()
max_rank = nomads_all_merged.loc[top_100_mask, 'rank'].max()

# Step 3: Create normalized score only for top 100
nomads_all_merged['nomad_web_score'] = 0  # Default to 0

nomads_all_merged.loc[top_100_mask, 'nomad_web_score'] = (
    (max_rank - nomads_all_merged.loc[top_100_mask, 'rank']) / (max_rank - min_rank) * 5
).round(1)
```

```python
min_val = nomads_all_merged['digital_nomads_count'].min()
max_val = nomads_all_merged['digital_nomads_count'].max()

nomads_all_merged['popularity_score'] = 1 + (
    (nomads_all_merged['digital_nomads_count'] - min_val) /
    (max_val - min_val)
) * 4

# Round to 1 decimal place
nomads_all_merged['popularity_score'] = nomads_all_merged['popularity_score'].round(1)
```

```python
min_val = nomads_all_merged['safety_value'].min()
max_val = nomads_all_merged['safety_value'].max()

nomads_all_merged['safety_score'] = 1 + (
    (nomads_all_merged['safety_value'] - min_val) /
    (max_val - min_val)
) * 4

# Round to 1 decimal place
nomads_all_merged['safety_score'] = nomads_all_merged['safety_score'].round(1)
```

```python
# Apply weights to each score
nomads_all_merged['coworking_potential_weighted_score'] = (
    nomads_all_merged['_score'] * 0.3 +
    nomads_all_merged['coworking_potential_score'] * 0.3 +
    nomads_all_merged['nomad_web_score'] * 0.2 +
    nomads_all_merged['popularity_score'] * 0.2
).round(2)
```

```python
# Assign rank (1 = best, higher score = better rank)
nomads_all_merged['coworking_rank'] = nomads_all_merged['coworking_potential_weighted_score'].rank(
    method='dense',  # No gaps in ranking
    ascending=False  # Higher score = better rank (rank 1 is the best)
).astype(int)  # Convert to integer

# Sort by rank (best first)
nomads_all_merged = nomads_all_merged.sort_values(by='coworking_rank')
```
