```python
trips = pd.read_csv('../data/raw/trips.csv', delimiter=',')
```

```python
trips = trips.drop(columns=["Unnamed: 0","country_slug","place_slug"])
```

```python
trips.rename(columns={'place':'city'},inplace=True)
```

```python
trips.head()
```

```python
# Convert date columns to datetime format
trips['date_start'] = pd.to_datetime(trips['date_start'], errors='coerce')
trips['date_end'] = pd.to_datetime(trips['date_end'], errors='coerce')
```

```python
# Drop rows where date conversion failed
trips = trips.dropna(subset=['date_start', 'date_end'])
```

```python
# Calculate travel duration
trips['day_travel'] = (trips['date_end'] - trips['date_start']).dt.days
```

```python
trips
```

```python
trips.dtypes
```

```python
# Reorder columns
new_column_order = ['username', 'country', 'country_code', 'city', 'date_start', 'date_end', 'day_travel', 'latitude', 'longitude']
trips = trips[new_column_order]
```

```python
trips
```

```python
from unidecode import unidecode

# Normalize address column
trips["city"] = trips["city"].apply(lambda x: unidecode(str(x)))
trips["country"] = trips["country"].apply(lambda x: unidecode(str(x)))
```

```python
trips.to_csv("../data/clean/sql_users_trips.csv", index=False, sep=";")
```
